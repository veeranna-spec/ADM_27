```python
data=pd.read_csv('titanic.csv')
print(data.isnull().sum())
```

```python
# Handle missing values (impute with mean, median, or mode)
data['Age'].fillna(data['Age'].mean(), inplace=True) # Impute with mean
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True) # Impute
```

```python
print("\nMissing values after imputation:")
print(data.isnull().sum())
```

```python
!pip install scikit-learn  # Install scikit-learn if you haven't already
from sklearn.preprocessing import MinMaxScaler # Import MinMaxScaler

scaler_minmax = MinMaxScaler()
columns_to_normalize = ['Age', 'Fare']
data[columns_to_normalize] = scaler_minmax.fit_transform(data[columns_to_normalize])
```

```python
# Import StandardScaler from sklearn.preprocessing
from sklearn.preprocessing import StandardScaler

scaler_standard = StandardScaler()
data[columns_to_normalize] = scaler_standard.fit_transform(data[columns_to_normalize])
```
